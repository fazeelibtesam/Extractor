```python
import os
import h5py

input_dir = '/content/sample_data/archive1'

for filename in os.listdir(input_dir):
    if filename.endswith('.mat'):
        mat_file_path = os.path.join(input_dir, filename)
        with h5py.File(mat_file_path, 'r') as mat_data:
            print(f"Contents of {filename}:")
            for key in mat_data.keys():
                print(f"Key: {key}, Type: {type(mat_data[key])}")
                if isinstance(mat_data[key], h5py.Group):
                    print("Subkeys:")
                    for subkey in mat_data[key].keys():
                        print(f"  Subkey: {subkey}, Type: {type(mat_data[key][subkey])}")
```

```python
import os
import h5py
import numpy as np
from PIL import Image
```

```python
input_dir = '/content/sample_data/archive1'
output_dir = '/content/sample_data/converted_images'
os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    if filename.endswith('.mat'):
        mat_file_path = os.path.join(input_dir, filename)
        with h5py.File(mat_file_path, 'r') as mat_data:
            cjdata = mat_data['cjdata']

            if 'image' in cjdata:
                image_data = cjdata['image'][()]
                if isinstance(image_data, np.ndarray):
                    if image_data.dtype != np.uint8:
                        image_data = (image_data - np.min(image_data)) / (np.max(image_data) - np.min(image_data)) * 255
                        image_data = image_data.astype(np.uint8)

                    img = Image.fromarray(image_data)

                    jpg_file_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}_image.jpg")
                    img.save(jpg_file_path)
                    print(f"Converted {filename} to {jpg_file_path}")

print("Conversion completed.")
```

```python
import shutil

# Define the directory containing the JPG images
output_dir = '/content/sample_data'

# Create a zip file of the images
shutil.make_archive(output_dir, 'zip', output_dir)

print("Zip file created.")
```
